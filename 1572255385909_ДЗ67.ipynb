Реализуйте свой класс Complex для комплексных чисел, аналогично встроенной реализации complex: 1. Добавьте конструктор класса 2. Реализуйте операции проверки на равенство, сложения, вычитания, произведения и деления комплексных чисел (eq, add, sub, mul, truediv) 3. Реализуйте операцию модуля (abs, вызываемую как |c|) 4. Оба класса должны давать осмысленный вывод как при print, так и просто при вызове в ячейке

p.s. обязательно использовать магические методы (см лабу)

```python
import math

class complexum:
    def __init__(self, re, im):
        self.re = re
        self.im = im
    
    def __eq__(self, other):
        if self.re == other.re and self.im == other.im:
            return True
        else:
            return False
      
    def __add__(self, other):
        return complexum(self.re + other.re, self.im + other.im)
    
    def __sub__(z1, z2):
        return complexum(z1.re - z2.re, z1.im - z2.im)
    
    def __mul__(z1, z2):
        return complexum((z1.re)*(z2.re)-(z1.im)*(z2.im), (z1.im)*(z2.re) + (z1.re)*(z2.im))
        
    def __truediv__(z1, z2):
        return complexum(((z1.re)*(z2.re)+(z1.im)*(z2.im))/((z2.re)*(z2.re)+(z2.im)*(z2.im)), ((z2.re)*(z1.im)-(z1.re)*(z2.im))/((z2.re)*(z2.re)+(z2.im)*(z2.im)))
    
    def __abs__(z):
        return(math.sqrt((z.re)*(z.re)+(z.im)*(z.im)))
    
    
    
z1=complexum(1, 2)
z2=complexum(1, 2)
print(z1.re, z1.im, z2.re, z2.im)
v=(z1==z2)
print(v)
```

Создать класс клетки с данными ДНК - последовательность символов. Клетка должна уметь делиться (написать метод класса). Реализовать процесс деления так, чтобы он отражал биологический процесс. Как делать - решать вам.

```python
import random

def bad_copy(b): #мутации при копировании генома
    a = list(b)
    copy = [None]*len(a)
    for i in range(len(a)):
        err = random.random()
        if err < 0.025:
            copy[i] = 'a'
        elif err < 0.05:
            copy[i] = 't'
        elif err < 0.075:
            copy[i] = 'g'
        elif err < 0.1:
            copy[i] = 'c'
        else:
            copy[i] = a[i]
    return ''.join(copy)
        
    

# dna = [[хроматида1хромосомыA, хроматида2хромосомыA], [хроматида1хромосомы, хроматида2хромосомыB]]
class cell:
    def __init__(self, dna):
        self.dna = dna
        
    def __repr__(self):
        print(self.dna)
        pass
    
    def __str__(self):
        return '{' + str(self.dna[0][0])+', '+str(self.dna[0][1])+'}, {'+str(self.dna[1][0])+', '+str(self.dna[1][1]) + '}'
    
    def replication(self):
        self.dna[0][1] = bad_copy(self.dna[0][0])
        self.dna[1][0] = bad_copy(self.dna[0][0])
        self.dna[1][1] = bad_copy(self.dna[0][0])
    
    def mitosis(self):        
        dna2=[[None]*2]*2
        dna2[0][0] = bad_copy(self.dna[0][0])
        dna2[1][0] = bad_copy(self.dna[1][0])
        self.dna[0][1] = None
        self.dna[1][1] = None
        return(self, cell(dna2))
    
    def meiosis(cell0):
        s1 = list(bad_copy(list(cell0.dna[0][0])))
        s2 = list(bad_copy(list(cell0.dna[0][1])))
        s3 = list(bad_copy(list(cell0.dna[1][0])))
        s4 = list(bad_copy(list(cell0.dna[1][1])))
        
        for i in range(len(s1)):
            q1 = random.random() # определяет, случится ли кроссинг-овер между первыми хроматидами
            q2 = random.random() # определяет, случится ли кроссинг-овер между вторыми хроматидами
            
            if q1 < 0.5:
                s1[i], s3[i] = s3[i], s1[i] # кроссинг-овер 1
            if q2 < 0.5:
                s2[i], s4[i] = s4[i], s2[i] # кроссинг-овер 2
        
        q1 = [[''.join(s1), None], [None, None]]
        q2 = [[''.join(s2), None], [None, None]]
        q3 = [[''.join(s3), None], [None, None]]
        q4 = [[''.join(s4), None], [None, None]]
        
        cell1 = cell(q1)
        cell2 = cell(q2)
        cell3 = cell(q3)
        cell4 = cell(q4)
        
        return(cell1, cell2, cell3, cell4)
    
        

dna_seryj = [['atgc', 'atgc'], ['atgc', 'atgc']] #ДНК Сереги
serGAY = cell(dna_seryj) # создаем клетку Сергея
print('serGAY: ', serGAY)
serGAY, anton = serGAY.mitosis() # Сережа поделился митозом и родил на свет Антоху - своего клона (с точностью до мутаций, определяемых err в bad_copy)
serGAY.replication()
anton.replication()
print('anton: ', anton)
print('serGAY after mitosis: ', serGAY)

dima, misha, slavik, lexa = anton.meiosis()
dima.replication()
misha.replication()
slavik.replication()
lexa.replication()

print('dima: ', dima)
print('misha: ', misha)
print('slavik: ', slavik)
print('lexa: ', lexa)

# можно поиграться с этим делом и увидет кучу всего интересного(нет)
# я очень долго пытался заставить это все заработать, поставьте побольше баллов за эту задачц, пж :3
```
