```python
class aircraft:
    def __init__(self):
        # собственные характеристики
        self.thrust = 10
        self.HP = 10
        self.maneuvr = 10
        self.radar_range = 10
        self.missile_range = 5
        self.missile_num = 2
        self.missile_dmg = 2
        self.frag = 0
        
        print("welcome")
        print("thrust = ", self.thrust)
        print("HP = ", self.HP)
        print("maneuvrability = ", self.maneuvr)
        print(" radar range = ", self.radar_range)
        print("missile range = ", self.missile_range)
        print("missile quantity = ", self.missile_num)
        print("missile damage = ", self.missile_dmg)
        print("frags: ", self.frag)
    
    
    # "дебаффы"
    def engine_hit(self, host_missile_dmg):
        self.HP -= host_missile_dmg        
        self.thrust -= host_missile_dmg
        
        print("engine hit")
        print("HP = ", self.HP, "thrust = ", self.thrust)
    
    def wing_hit(self, host_missile_dmg):
        self.HP -= host_missile_dmg
        self.maneuvr -= host_missile_dmg
        self.missile_num -= 1
        
        print("wing hit")
        print("HP = ", self.HP,"maneuvrability = ", self.maneuvr, "missile quantity = ", self.missile_num)
        
    def radar_hit(self, host_missile_dmg):
        self.HP -= host_missile_dmg
        self.radar_range -= host_missile_dmg
        
        print("radar hit")
        print("HP = ", self.HP, "radar range = ", self.radar_range)
    
    # атака     
    def enemyhit(self):
        self.frag += 1
        self.missile_num -= 1
        
        print("enemy shot down")
        print("frags: ", self.frag)
    
    def missed(self):
        self.missile_num -= 1
        
        print("miss")
        
    
    # "баффы"
    def aircraft_lvlup(self):
        self.HP += 10
        
        print("aircraft level up")
        print("HP = ", self.HP)
        
    def engine_lvlup(self, lvlup):
        self.thrust += lvlup
        
        print("engine level up")
        print("thrust = ", self.thrust)
        
    def wing_lvlup(self, lvlup):
        self.missile_num += 2
        self.maneuvr += lvlup
        
        print("wing level up")
        print("missile number = ", self.missile_num, " maneuvr = ", self.maneuvr)
        
    def missile_lvlup(self, lvlup):
        self.missile_range += lvlup
        self.missile_dmg += 2
        
        print("missile level up")
        print("missile range = ", self.missile_range, " missile damage = ", self.missile_dmg)
        
    def radar_lvlup(self, lvlup):
        self.radar_range += lvlup
        
        print("radar level up")
        print("radar range = ", self.radar_range)
        
    def missile_stock(self):
        self.missile_num += 2
        
        print("missile stock renewed")
        print("missile quantity = ", self.missile_num)
        
        
        
        
        
# пример "воздушного боя"

G = aircraft() # инстанциация

G.enemyhit() # противник сбит
G.missed()   # мимо
G.missile_stock() # "подобрал" боеприпасы
G.wing_hit(2) # "ранен", повреждено крыло на 2
G.wing_lvlup(4) # "вылечил" крыло на 4
G.missed() # не попал
G.enemyhit() # попал
        
        
    
    
        
        
```
